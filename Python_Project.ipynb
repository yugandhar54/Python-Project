```python
import pandas as pd
data = pd.read_csv(r"C:\Users\ASUS\Downloads\py_1\Candy_Sales.csv")
mean_sales = data['Sales'].mean()
print("Mean Sales:", mean_sales)
median_sales = data['Sales'].median()
print("Median Sales:", median_sales)
mode_sales = data['Sales'].mode()
print("Mode Sales:", mode_sales)
total_sales = data['Sales'].sum()
print("Total Sales:", total_sales)
sales_count = data['Sales'].count()
print("Total Number of Sales Records:", sales_count)
sales_range = data['Sales'].max() - data['Sales'].min()
print("Range of Sales:", sales_range)
```

```python
max_sales = data['Sales'].max()
print("Maximum Sales:", max_sales)
min_sales = data['Sales'].min()
print("Minimum Sales:", min_sales)
std_dev_sales = data['Sales'].std()
print("Standard Deviation of Sales:", std_dev_sales)
variance_sales = data['Sales'].var()
print("Variance of Sales:", variance_sales)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv(r"C:\Users\ASUS\Downloads\py_1\Candy_Sales.csv")
data.columns = data.columns.str.strip()
print(data.columns)
data['Order Date'] = pd.to_datetime(data['Order Date'], dayfirst=True)
sales_trends = data.groupby('Order Date')['Sales'].sum()
plt.figure(figsize=(10, 6))
sales_trends.plot(kind='line', marker='o')
plt.title("Candy Sales Trends Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(data['Price'], data['Sales'], alpha=0.7, color='green', edgecolor='k')
plt.title("Scatter Plot: Price vs Sales")
plt.xlabel("Price")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
data.columns = data.columns.str.strip()  
np.random.seed(42)  
campaigns = ['Campaign A', 'Campaign B', 'Campaign C']
data['Campaign'] = np.random.choice(campaigns, size=len(data))
campaign_sales = data.groupby('Campaign')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(8, 6))
campaign_sales.plot(kind='bar', color='orange')
plt.title("Effectiveness of Marketing Campaigns")
plt.xlabel("Campaign")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
data.columns = data.columns.str.strip() 
data['Order Date'] = pd.to_datetime(data['Order Date'], dayfirst=True, errors='coerce')
data['Month'] = data['Order Date'].dt.strftime('%B')  
data['Month_Num'] = data['Order Date'].dt.month  
monthly_sales = data.groupby(['Month_Num', 'Month'])['Sales'].sum().sort_index()
monthly_sales.index = monthly_sales.index.droplevel(0)
plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='bar', color='skyblue')
plt.title("Monthly Inventory Needs")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
